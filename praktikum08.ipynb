**1. Import Library**

```python
import matplotlib.pyplot as plt
import numpy as np
```

**2. Dataset**

```python
# data penjualan smartphone
brands = ['Samsung', 'Iphone', 'Vivo', 'Oppo']
sales_2023 = [500, 700, 300, 400]
sales_2024 = [550, 750, 350, 450]
sales_2025 = [600, 800, 400, 500]
```

**3. Grafik Basic (Pie Chart dan Area Chart)**

```python
y_pos = np.arange(len(brands))

# ====================================
# 1. BASIC HORIZONTAL BAR CHART (2023)
# ====================================
fig1, ax1 = plt.subplots()
ax1.barh(brands, sales_2023)
ax1.set_title('Horizontal Bar Chart - 2023')
ax1.set_xlabel('Jumlah Penjualan')
ax1.set_ylabel('Merk')

plt.show()

# =================================================
# 2. BASIC STACKED HORIZONTAL BAR CHART (2023 & 2024)
# =================================================
fig2, ax2 = plt.subplots()
ax2.barh(brands, sales_2023, label='2023')
ax2.barh(brands, sales_2024, left=sales_2023, label='2024')

ax2.set_title('Penjualan Smartphone Stacked (2023 & 2025)')
ax2.set_xlabel('Jumlah Penjualan')
ax2.set_ylabel('Merk')
ax2.legend()

plt.show()
```

**4. Grafik Kustomisasi (Pie Chart dan Area Chart)**

```python
y_pos = np.arange(len(brands))
colors = ['gold', 'skyblue', 'coral', 'lightgreen']

# 1. CUSTOM HORIZONTAL BAR CHART (2023)
fig3, ax3 = plt.subplots()
ax3.barh(brands, sales_2023, color=colors, edgecolor='black', height=0.6)
ax3.set_title('Penjualan Smartphone Tahun 2023 (Customized)')
ax3.set_xlabel('Jumlah Penjualan')
ax3.set_ylabel('Merk')
ax3.grid(axis='x', linestyle='--', alpha=0.7)

# menampilkan label nilai di bar chart
for i, v in enumerate(sales_2023):
    ax3.text(v + 10, i, str(v), va='center')
plt.show()

# 2. CUSTOM STACKED HORIZONTAL BAR CHART (20232025)
fig4, ax4 = plt.subplots()
ax4.barh(brands, sales_2023, color='lightcoral', edgecolor='black',label='2023')
ax4.barh(brands, sales_2024, left=sales_2023, color='lightskyblue',  edgecolor='black', label='2024')

ax4.set_title('Penjualan Smartphone Tahun 2023 (Stacaked - Customized)')
ax4.set_xlabel('Jumlah Penjualan')
ax4.set_ylabel('Merk')
ax4.legend()
ax4.grid(axis='x', linestyle='--', alpha=0.5)

# menampilkan label nilai di bar chart
total_sales = np.array(sales_2023) + np.array(sales_2024) + np.array(sales_2025)
for i, v in enumerate(total_sales):
    ax4.text(v + 10, i, str(v), va='center')
plt.show()
```

**5. Grafik Multiple (Pie Chart dan Area Chart)**

```python
y_pos = np.arange(len(brands))

# ===============================================
# 1. MULTIPLE HORIZONTAL BAR CHART (2023 vs 2024)
# ===============================================
fig5, ax5 = plt.subplots()
ax5.barh(y_pos - 0.2, sales_2023, color='skyblue', height=0.4, label='2023')
ax5.barh(y_pos + 0.2, sales_2024, color='salmon', height=0.4, label='2024')

ax5.set_yticks(y_pos, brands)
ax5.set_title('Penjualan Smartphone 2023 vs 2024')
ax5.set_xlabel('Jumlah Penjualan')
ax5.set_ylabel('Merk')
ax5.legend()

plt.show()

# ====================================================
# 2. MULTIPLE STACKED HORIZONTAL BAR CHART (20232025)
# ====================================================
fig6, ax6 = plt.subplots()
ax6.barh(y_pos, sales_2023, color='skyblue', label='2023')
ax6.barh(y_pos, sales_2024, left=sales_2023, color='salmon', label='2024')
ax6.barh(y_pos, sales_2025, left=np.array(sales_2023) + np.array(sales_2024), color='lightgreen', label='2025')

ax6.set_yticks(y_pos, brands)
ax6.set_title('Penjualan Smartphone 2023 vs 2025 (Stacked)')
ax6.set_xlabel('Jumlah Penjualan')
ax6.set_ylabel('Merk')
ax6.legend()

plt.show()
```
